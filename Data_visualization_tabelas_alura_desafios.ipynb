```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/loja_livro_filmes.csv')
df.head()
```

```python
df.info()
```

```python
# Transformando o tipo da coluna data_pedido de object para date_time
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
df['data_chegada'] = pd.to_datetime(df['data_chegada'])
df.info()
```

##Desafio 1:
O primeiro desafio é construir uma visualização que permita identificar quais países mais solicitaram produtos por pedidos para que assim possa ser feito um estudo quanto a distribuição e logística de produtos.

```python
df['pais'].unique()
```

```python
df_pedidos_paises = df.groupby(['pais'])['unidades'].sum().nlargest(9).copy()
df_pedidos_paises = df_pedidos_paises.reset_index()
df_pedidos_paises
```

```python
# Renomeado as colunas
df_pedidos_paises.columns = ['País', 'Unidades pedidas']

# Criando a coluna de ranking
df_pedidos_paises['Ranking'] = df_pedidos_paises.index + 1
df_pedidos_paises.set_index('Ranking', inplace=True)
df_pedidos_paises
```

```python
# Criando a visualização
s = df_pedidos_paises.style
s
```

```python
s.format({'Unidades pedidas': '{} produtos'})
s
```

##Desafio 2:
O objetivo desse desafio é construir uma visualização que permita identificar quais categorias de produto geraram mais e menos custos para a loja.



```python
df.head()
```

```python
df_custos = df.groupby(['categoria'])[['custo_produto']].sum().copy()
df_custos.index.name = 'Categoria'
df_custos.columns = ['Custo de pedidos']
df_custos
```

```python
estilo_custos.format({'Custo de pedidos':'R$ {:,.2f}'})\
                     .highlight_max(color='green')\
                     .highlight_min(color='darkred')
estilo_custos
```

```python
cabecalho = {
    'selector':'th',
    'props':'text-align: center; font-family: arial; font-weight: bold;'
}

estilo_custos.set_table_styles([cabecalho], overwrite=False)
estilo_custos
```

##Desafio 3:
O desafio é construir uma visualização que permita informar a quantidade e distribuição de pedidos por tipo de desconto.


```python
df.info()
```

```python
df_desconto = pd.DataFrame(df['tipo_desconto'].value_counts())
df_desconto.columns = ['N° de pedidos']
df_desconto.index.name = 'Tipo de desconto'
df_desconto
```

```python
porcentagem = df_desconto['N° de pedidos'].to_numpy()
porcentagem = 100 * porcentagem/porcentagem.sum()
porcentagem
```

```python
df_desconto['Porcentagem de tipo de desconto'] = porcentagem
df_desconto
```

```python
estilo_desconto = df_desconto.style
estilo_desconto
```

```python
cabecalho ={
    'selector':'th',
    'props':'text-align: left; font-family: Arial; font-weight: bold; background-color: #4682B4; color: #FFD700;'
}

celulas = {
    'selector':'td',
    'props':'background-color: #4682B4; color:#FFF5EE;'
}

estilo_desconto.set_table_styles([cabecalho, celulas])
```

```python
estilo_desconto.format({'Porcentagem de tipo de desconto':'{:.2f}%'})\
               .bar(subset='Porcentagem de tipo de desconto', vmin=0, vmax=100, color='green', height=70, width=60)
```

##Desafio 4:
Nosso desafio agora é fornecer uma visualização que permita **mostrar o tempo médio de entrega de pedidos durante os meses dos anos de 2013 a 2015 por cada país.**

```python
df.head()
```

```python
df_datas = df.copy()
df_datas = df_datas.sort_values('data_pedido')
df_datas = df_datas.reset_index(drop=True)
df_datas
```

```python
df_datas = df_datas[df_datas['data_pedido'] >= '2013-01-01']
df_datas
```

```python
df_datas['meses'] = df_datas['data_pedido'].dt.strftime('%Y - %b')
df_datas
```

```python
tempo_demora = (df_datas['data_chegada'] - df_datas['data_pedido']).dt.days
df_datas['tempo_entrega'] = tempo_demora
df_datas
```

```python
entregas = df_datas.pivot_table(index='pais', columns='meses', values='tempo_entrega', aggfunc='mean', sort=False)
```

```python
estilo_entregas = entregas.style.format('{:,.2f}')
estilo_entregas.set_sticky(axis='index')
estilo_entregas
```

##Desafio 5:
Nosso desafio envolve uma visualização que permita separar os gastos de envio de cada produto (custo total) por tipo de cliente, de modo que seja possível entender o tipo de produto que mais gerou custos de envio em cada tipo de cliente. Além disso, é interessante também mostrar a distribuição desses custos por país enviado, permitindo entender também os gastos por país.



```python
## Preparar os dados

# Criar um DataFrame alternativo para ter um nome de coluna coerente com a visualização
df_alteracao = df.rename(columns={'pais': 'Custo por país ($)'})

# Agrupar os dados de tipo de cliente, com a categoria de produtos e os países que esses clientes importam o produto
df_clientes = df_alteracao.pivot_table(index=['tipo_cliente', 'categoria'], columns = 'Custo por país ($)', values = 'custo_produto', aggfunc = 'sum')

# Renomear o cabeçalho para nomes mais coerentes
df_clientes.rename_axis(index={'tipo_cliente': 'Tipo cliente', 'categoria': 'Categoria produto'}, inplace = True)

# Adicionar uma coluna que mostra a soma dos custos por tipo de cliente e categoria de produto
df_clientes['Total ($)'] = df_clientes.sum(axis=1)

## Criar a visualização
s_clientes = df_clientes.style.format('{:,.2f}')

# Criar os estilos base para a tabela, a partir da padronização das fontes, plano de fundo para branco e o alinhamento do cabeçalho para o topo
tabela = {
    'selector': 'td, th',
    'props': 'font-weight: normal; font-family: Arial; text-align: right; background-color: white; color:black;'
}
# Alinhar verticalmente o cabeçalho para o topo permite a pessoa observadora identificar onde começa a hierarquização dos grupos B2B e B2C
cabecalho = {
    'selector': 'th',
    'props': 'vertical-align: top'
}
s_clientes.set_table_styles([tabela, cabecalho])

# Adicionar linha de separação do cabeçalho superior
# Destacar as categorias que mais geraram custo com a aplicação de negrito nas fontes
s_clientes.set_table_styles({
    ('B2B', 'BlueRay'): [{'selector': 'th', 'props': 'border-top: 1px solid #181818'},
              {'selector': 'td', 'props': 'border-top: 1px solid #181818'}],
    ('B2B', 'Livro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
    ('B2C', 'Livro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
}, overwrite=False, axis=1)

# Destacar os elementos em cada tipo de cliente que mais geraram custos com uma seleção
s_clientes.set_table_styles({
    'Total ($)': [{
        'selector': '.true',
        'props': 'font-weight: bold'
    }]
},overwrite=False,axis=0)

cores_coluna = pd.DataFrame(['false','false','false','true','false','false','false','true'],index= df_clientes['Total ($)'].index,
                            columns = ['Total ($)'])
# Visualização final
s_clientes.set_td_classes(cores_coluna)
```
